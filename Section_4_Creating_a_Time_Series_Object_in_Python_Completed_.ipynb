# Fill the NaN values
-------------------------
This code fills the missing values (NaN) in the 'spx' column of the df_comp DataFrame using the forward-fill method (method='ffill').

This means that each missing value is replaced with the last known valid value before it.

```python
df_comp.isna().sum()
```

```python
df_comp.ftse=df_comp.ftse.fillna(method='bfill')
```

```python
df_comp.dax=df_comp.dax.fillna(value = df_comp.dax.mean())
```

# Fill the NaN values with the average of the column
----------------------
This code fills the missing values (NaN) in the 'dax' column of the df_comp DataFrame with the mean of the existing values in the 'dax' column.

This is an alternative method for handling missing data compared to forward-fill or backward-fill.

```python
# Find indices of NaN values in the 'dax' column
nan_indices = df_comp[df_comp['dax'].isna()].index

# Get the index of the first NaN value
first_nan_index = nan_indices[0]

# Get the index before the first NaN value
index_before_nan = df_comp.index[df_comp.index.get_loc(first_nan_index) - 1]

# Get the value at the index before the first NaN value
value_before_nan = df_comp.loc[index_before_nan, 'dax']

print(f"The first NaN value in the 'dax' column is at index: {first_nan_index}")
print(f"The index before the first NaN is: {index_before_nan}")
print(f"The value before the first NaN is: {value_before_nan}")
```

This code helps to understand how the ffill method works by specifically finding the value right before the first missing value in the 'dax' column.

Here's a breakdown:

df_comp[df_comp['dax'].isna()].index: This part finds all the index values where the 'dax' column has NaN values.

nan_indices[0]: This selects the first index from the list of NaN indices.
df_comp.index.get_loc(first_nan_index) - 1: This gets the integer position of the first NaN index and then subtracts 1 to get the integer position of the index immediately before it.
df_comp.index[...]: This uses the integer position found in the previous step to get the actual index value.
df_comp.loc[index_before_nan, 'dax']: This uses the index value found in the previous step to retrieve the value from the 'dax' column at that specific index.
Finally, the code prints the first NaN index, the index before it, and the value at that preceding index. This demonstrates what value would be used to fill the first NaN if ffill() were applied to the 'dax' column.
